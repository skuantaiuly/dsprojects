```python
import pandas as pd
import seaborn as sb; sb.set()
import numpy as np
%matplotlib inline
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
data
```

```python
x = list(data['x'])
y = list(data['y'])
color = list(data['color'])
```

```python
plt.scatter(x,y,s=20)
plt.show()
```

```python
arr2d = np.column_stack([x,y])
arr2d
```

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(arr2d)
y_kmeans = kmeans.predict(arr2d)
```

```python
plt.figure(figsize=(8,6))
plt.xlabel('x')
plt.ylabel('y')
plt.title('Clustered data (k=3)')
plt.scatter(arr2d[:, 0], arr2d[:, 1], c = color, s = 50, cmap='Set1')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c ='black', s=100 )
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.xlabel('x')
plt.ylabel('y')
plt.title('Clustered data (k=3)')
plt.scatter(arr2d[:, 0], arr2d[:, 1], c = y_kmeans, s = 50, cmap='Set1')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c ='black', s=100 )
plt.show()
```
